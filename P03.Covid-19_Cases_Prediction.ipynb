```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sn
```

```python
cases = pd.read_csv("DataSets/CONVENIENT_global_confirmed_cases.csv")
deaths = pd.read_csv("DataSets/CONVENIENT_global_deaths.csv")
```

```python
cases = cases.dropna()
deaths = deaths.dropna()
```

```python
cases.iloc[0,1]
```

```python
for i in range(1,280):
    cases[cases.columns[i]] = pd.to_numeric(cases[cases.columns[i]], downcast="float").astype(float)
    deaths[deaths.columns[i]] = pd.to_numeric(deaths[deaths.columns[i]], downcast="float").astype(float)
```

```python
# This is the array of daily cases
total_cases = []
total_deaths = []
for i in range(0,len(cases)):
    total_cases.append(sum(cases.iloc[i,1:].values))
    total_deaths.append(sum(deaths.iloc[i,1:].values))
```

```python
# daily cases
plt.figure(figsize=(12,6))
plt.plot(deaths['Country/Region'],total_deaths)
plt.plot(cases['Country/Region'],total_cases) 
```

```python
cases_month = []
deaths_month = []
cases_30 = []
deaths_30 = []
for i in range(1,19):
    cases_month.append(cases['Country/Region'][i*30])
    deaths_month.append(deaths['Country/Region'][i*30])
    cases_30.append(total_cases[i*30])
    deaths_30.append(total_deaths[i*30])
```

```python
plt.figure(figsize=(12,6))
plt.plot(deaths_month,deaths_30)
plt.plot(cases_month,cases_30) 
```
