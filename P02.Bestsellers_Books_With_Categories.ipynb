```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv("DataSets/bestsellers with categories.csv")
df.head()
```

```python
len(df.Name.unique())
df = df.drop_duplicates('Name')
```

```python
df.Name.unique()
```

```python
df.groupby("Genre").count()
```

```python
df.isna().sum()
```

```python
df.Year.max()
```

```python
df['Years'] = 2020 - df['Year']
df['Years'] = df['Years']/df['Years'].max()
df['User Rating'] = df['User Rating']/df['User Rating'].max()
df['Reviews'] = df['Reviews']/df['Reviews'].max()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Encoded_Genre = le.fit_transform(df.Genre)
df['Encoded_Genre'] = Encoded_Genre
```

```python
df
```

```python
y1 = np.arange(2009,2014)
y2 = np.arange(2014,2020)
g_count = df['Encoded_Genre'].value_counts()
fig,ax = plt.subplots(2,6,figsize=(12,6))
ax[0,0].pie(x=g_count.values,autopct='%1.1f%%')
ax[0,0].set_title("Over all")
for i,year in enumerate(y1):
    counts = df[df['Year'] == year]['Encoded_Genre'].value_counts()
    ax[0,i+1].pie(x = counts.values,autopct='%1.1f%%')
    ax[0,i+1].set_title(year)
for i,year in enumerate(y2):
    counts = df[df['Year'] == year]['Encoded_Genre'].value_counts()
    ax[1,i].pie(x = counts.values,autopct='%1.1f%%')
    ax[1,i].set_title(year)

fig.legend(g_count.index)
fig.show()
```

```python
df["Popularity"] = (df['Reviews'] * df['User Rating'])/df['Years']
df["Popularity"] = df["Popularity"]/df["Popularity"].max()
```

```python
df
```

```python
df = df.sort_values(by=["Encoded_Genre","Popularity"],ascending = False)
```

```python
non_fiction = df[df.Encoded_Genre == 1][['Popularity','Name','Author']].head(10)
fiction = df[df.Encoded_Genre == 0][['Popularity','Name','Author']].head(10)
```

```python
#Top 50 books
df1 = pd.DataFrame(non_fiction)
df2 = pd.DataFrame(fiction)
```

```python
plt.figure(figsize=(36,18))
plt.barh(df2.Author,df2.Popularity )
plt.barh(df1.Author,df1.Popularity )
plt.show()
```
